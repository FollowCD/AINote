### Windows
 Windows TensorFlow 0.12.1Python 3.5Pillow 3.4.2scipy 0.18.1  numpy 1.11.2 Miniconda 
```
conda create -n style-transfer python=3.5
activate style-transfer
pip install tensorflow
conda install scipy pillow
```
 (activate style-transfer)  TensorFlowScipy  Pillow

### OS X  Linux
 OS X  Linux TensorFlow 0.11.0Python 2.7.9Pillow 3.4.2scipy 0.18.1  numpy 1.11.2.


```
conda create -n style-transfer python=2.7.9
source activate style-transfer
pip install tensorflow
conda install scipy pillow
```
 (source activate style-transfer)  TensorFlowScipy  Pillow
